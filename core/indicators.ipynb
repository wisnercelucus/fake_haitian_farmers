```python
file_root = './inputs/'
```

```python
file_path = file_root + 'indicators.xlsx'
```

```python
# Load the Excel file
excel_data = pd.ExcelFile(file_path)

# Dictionary to store DataFrames for each sheet
dataframes = {}

# Loop through each sheet name
for sheet_name in excel_data.sheet_names:
    # Read the sheet into a DataFrame
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    # Store the DataFrame in the dictionary
    dataframes[sheet_name] = df
    print(f"DataFrame for sheet '{sheet_name}' created successfully!")

# Optionally display the DataFrames
for sheet_name, df in dataframes.items():
    print(f"\nDataFrame for sheet: {sheet_name}")
    print(df)
```
